## What is Jupyter?

Jupyter Notebook is an open-source web application that allows you to create and share documents that contain:

- Live code
- Visualizations
- Narrative text
- Equations
- Images
- Videos

Jupyter notebooks are widely used in data science, scientific computing, machine learning, and education.

## Cell Types

Jupyter notebooks contain cells of different types:

1. **Code cells**: Contain code that can be executed
2. **Markdown cells**: Contain formatted text (like this cell)
3. **Raw cells**: Contain content that is not processed by the notebook

Let's look at examples of each.

### Code Cells

Code cells contain code that can be executed. Below is a simple Python code cell:

````python
# This is a code cell
print("Hello, Jupyter!")

# Simple calculation
result = 5 * 10
print(f"5 * 10 = {result}")
````

### Markdown Cells

Markdown cells allow you to format text using [Markdown syntax](https://www.markdownguide.org/basic-syntax/).

You can create:

- **Bold text** with `**double asterisks**`
- *Italic text* with `*single asterisks*`
- Lists (like this one)
- [Links](https://jupyter.org) with `[text](URL)`
- Headers with `#` symbols (more `#` means smaller headers)
- Code blocks with triple backticks ``` ```
- Math equations using LaTeX: $E = mc^2$

## Common Operations in Jupyter

### 1. Working with Variables

````python
# Variables persist between cells
x = 10
y = 20
print(f"x = {x}, y = {y}")
````

````python
# Use variables from previous cells
z = x + y
print(f"x + y = {z}")
````

### 2. Data Visualization

Jupyter notebooks are excellent for data visualization. Here's an example using matplotlib:

````python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Enable inline plotting
%matplotlib inline

# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.title('Sine Wave', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('sin(x)', fontsize=14)
plt.grid(True)
plt.show()
````

### 3. Working with Data using Pandas

````python
# Import pandas
import pandas as pd

# Create a sample dataframe
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago', 'Boston'],
    'Salary': [70000, 80000, 90000, 100000, 110000]
}

df = pd.DataFrame(data)
df
````

````python
# Basic data exploration
print("DataFrame shape:", df.shape)
print("\nDataFrame info:")
df.info()
print("\nDataFrame statistics:")
df.describe()
````

````python
# Filtering data
df[df['Age'] > 30]
````

````python
# Visualization with pandas
ax = df.plot.bar(x='Name', y='Salary', figsize=(10, 6), title='Salary by Person')
ax.set_ylabel('Salary ($)')
````

### 4. Magic Commands

Jupyter notebooks support special commands called "magic commands" that start with `%` (line magics) or `%%` (cell magics).

````python
# List all available magic commands
%lsmagic
````

````python
# Timing execution
%time sum(range(1000000))
````

````python
# Cell magic to time the whole cell
%%time
result = 0
for i in range(1000000):
    result += i
print(f"Result: {result}")
````

### 6. Interactive Widgets

Jupyter supports interactive widgets using `ipywidgets`:

````python
# Import widgets
import ipywidgets as widgets
from IPython.display import display

# Create a slider
slider = widgets.IntSlider(
    value=10,
    min=0,
    max=100,
    step=1,
    description='Value:',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

# Display the slider
display(slider)
````

````python
# Interactive function
def update_plot(n=10):
    x = np.linspace(0, 10, 100)
    y = np.sin(n * x)
    
    plt.figure(figsize=(10, 6))
    plt.plot(x, y)
    plt.title(f'sin({n}x)')
    plt.grid(True)
    plt.show()

# Create interactive widget
widgets.interact(update_plot, n=(1, 10, 0.5))
````

## Keyboard Shortcuts

Jupyter has two modes: **Command mode** and **Edit mode**.

### Command Mode (press Esc to activate)
- **A**: Insert cell above
- **B**: Insert cell below
- **DD**: Delete cell
- **M**: Change cell to Markdown
- **Y**: Change cell to Code
- **Shift+Enter**: Run cell and select next
- **Ctrl+Enter**: Run cell and stay

### Edit Mode (press Enter to activate)
- **Shift+Enter**: Run cell and select next
- **Ctrl+Enter**: Run cell and stay
- **Alt+Enter**: Run cell and insert below
- **Esc**: Enter command mode

## Tips for Effective Jupyter Usage

1. **Name your notebooks** descriptively
2. **Use meaningful headings** to organize your work
3. **Restart and run all** periodically to ensure reproducibility
4. **Document your code** with markdown cells
5. **Use version control** (e.g., Git) to track changes
6. **Export your notebooks** for sharing (HTML, PDF, Python script)
7. **Create a table of contents** for large notebooks
8. **Clean outputs** before committing to version control
9. **Use Jupyter extensions** to enhance functionality
